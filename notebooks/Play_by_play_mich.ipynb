```python
import requests
import pandas as pd

pbp = []

for year in range(2001, 2020):
    for week in range(1,14):
        try:
            response = requests.get("https://api.collegefootballdata.com/plays?seasonType=regular&year={}&week={}&team=michigan".format(year, week))
            print("year={}&week={}".format(year,week),response)
            data_json = response.json()
            pbp.extend(data_json)
        except requests.exceptions.ConnectionError:
            print("year={}&week={}  Connection refused").format(week, year)
```

```python
df = pd.DataFrame(pbp)
df.iloc[0:10]
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = df[df['offense'] == 'Michigan']
scoring = df[df['scoring'] == True]
not_scoring = df[df['scoring'] == False]
sns.kdeplot(scoring['yards_gained'], label = 'Scoring Drives')
sns.kdeplot(not_scoring['yards_gained'], label = 'Non-Scoring Drives')
plt.xlim(-20, 100)
plt.title('Distributions of Yards Gained per Play on Scoring and Non-Scoring Drives')
plt.xlabel('Yards Gained')
plt.ylabel('Density')
```

```python
import numpy as np

first = df[df['down'] == 1]
second = df[df['down'] == 2]
third = df[df['down'] == 3]
# avg distance, avg yards gained, proportion of run/pass
first_data = []
second_data = []
third_data = []

first_data.append(first['distance'].mean())
second_data.append(second['distance'].mean())
third_data.append(third['distance'].mean())

first_data.append(first['yards_gained'].mean())
second_data.append(second['yards_gained'].mean())
third_data.append(third['yards_gained'].mean())

first_data.append(first['play_type'].str.contains('Rush').sum()*10/len(first))
second_data.append(second['play_type'].str.contains('Rush').sum()*10/len(second))
third_data.append(third['play_type'].str.contains('Rush').sum()*10/len(third))

ax = plt.subplot(111)
barWidth = .2
r1 = np.arange(len(first_data))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
plt.bar(r1, first_data, color='#7f6d5f', width=barWidth, edgecolor='white', label='First Down')
plt.bar(r2, second_data, color='#557f2d', width=barWidth, edgecolor='white', label='Second Down')
plt.bar(r3, third_data, color='#2d7f5e', width=barWidth, edgecolor='white', label='Third Down')
plt.xlabel('Metric', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(first_data))], ['Average Distance to 1st', 'Average Yards Gained', 'Rel. Rush Proportion'])
plt.legend()
plt.title('Breakdown of Metrics for Different Downs')
plt.show()

```
