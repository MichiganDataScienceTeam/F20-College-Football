```python
import requests
import pandas as pd

pbp = []

for year in range(2001, 2020):
    for week in range(1,14):
        try:
            response = requests.get("https://api.collegefootballdata.com/plays?seasonType=regular&year={}&week={}&team=michigan".format(year, week))
            print("year={}&week={}".format(year,week),response)
            data_json = response.json()
            pbp.extend(data_json)
        except requests.exceptions.ConnectionError:
            print("year={}&week={}  Connection refused").format(week, year)
```

```python
df = pd.DataFrame(pbp)
df.head()
```
