```python
import requests
import json
import pandas as pd
```

```python
# only gets the fbs teams
response = requests.get("https://api.collegefootballdata.com/teams/fbs")
print(response)
data = response.json()
teams = [i['school'] for i in data]
print(len(teams))
```

```python
# getting all data -> regular and postseason
play_by_play = []

# starting from 2005 -> data was sparse before then + 2005 is a conference realignment year
for team in teams:
    for year in range(2005, 2010):
        for week in range(1, 14):
            try:
                query = {'seasonType': "regular", 'year': year, 'week': week, 'team': team}
                response = requests.get("https://api.collegefootballdata.com/plays", params = query)
                print("team={}&year={}&week={}".format(team, year, week),response)
                json_data = response.json()
                play_by_play.extend(json_data)
            except requests.exceptions.ConnectionError as e:
                continue
```

```python
# creating csv file
df = pd.DataFrame(play_by_play)
df.to_csv("../data/play_by_play_fbs_regular_1.csv")
```
