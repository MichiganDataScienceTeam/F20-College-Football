```python
from elo import Rating, quality_1vs1, rate_1vs1
```

```python
teams = {}
for team in df.away_id.unique():
    teams[team] = Rating(1500)

for team in df.home_id.unique():
    if team in teams.keys():
        pass
    else:
        teams[team] = Rating(1500)
```

```python
teams
```

```python
for index, game in df.iterrows():
    if game['home_points'] > game['away_points']:
        teams[game['home_id']], teams[game['away_id']] = rate_1vs1(teams[game['home_id']], teams[game['away_id']])
    elif game['home_points'] < game['away_points']:
        teams[game['away_id']], teams[game['home_id']] = rate_1vs1(teams[game['away_id']], teams[game['home_id']])
    else:
        teams[game['home_id']], teams[game['away_id']] = rate_1vs1(teams[game['home_id']], teams[game['away_id']], drawn=True)
```
