```python
import requests
import pandas as pd
```

```python
drive = []

for year in range(2005, 2021):
    response = requests.get('http://api.collegefootballdata.com/drives', params={'year':'{}'.format(year)})
    
    drive_json = response.json()
    drive.extend(drive_json)
```

```python
df = pd.DataFrame(drive)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.to_csv("../data/drive_data.csv")
```

```python
umdrive =[]
for year in range (2016,2019):
    umresponse = requests.get('http://api.collegefootballdata.com/drives?offense=Michigan&defense=Minnesota', params={'year':'{}'.format(year)})
    umdrive_json = umresponse.json()
    umdrive.extend(umdrive_json)
```

```python
#"https://api.collegefootballdata.com/drives?seasonType=regular&year=2019&offense=Michigan"
umdf = pd.DataFrame(umdrive)
umdf.head()
```
