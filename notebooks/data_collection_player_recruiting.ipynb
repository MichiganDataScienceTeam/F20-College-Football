```python
import requests

recruitTypes = ["HighSchool", "JUCO", "PrepSchool"]
player_recruit = []

for year in range(2000,2021):
    for recruitType in recruitTypes:
        response = requests.get("https://api.collegefootballdata.com/recruiting/players?year={}&classification={}".format(year,recruitType))
        print("year={}&classification={}".format(year,recruitType),response) # <Response [200]> indicates successful operation 
        data_json = response.json()
        player_recruit.extend(data_json)
```

```python
len(player_recruit)
```

```python
import pandas as pd
player_recruit_df = pd.DataFrame(player_recruit)
player_recruit_df.head()
```
